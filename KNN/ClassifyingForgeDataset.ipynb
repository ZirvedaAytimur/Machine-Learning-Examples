##### Making an algorithm with scikit-learn

```python
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
#fit the classifier using training set
clf.fit(X_train, y_train)
```

```python
#for each data points in the test set, this computes its nearest neighbours in the training set,
#this computes its nearest neighbours in the training set
#and finds the most common class
print("Test set predictions: {}".format(clf.predict(X_test)))
```

```python
print("Test set accuracy: {:.2f}".format(clf.score(X_test, y_test)))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 3))

for n_neighbours, ax in zip([1, 3, 9], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbours).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{} neighbour(s)".format(n_neighbours))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")

axes[0].legend(loc=3)
```
